# The Python script with codes for Analysis and Data Viz.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:\\Users\HP\\Downloads\\Dataset\\Train_data.csv')
data
```

```python
print("Dataset Overview:")
print(data.head())  # Display the first few rows of the dataset
print("\nDataset Info:")
print(data.info())  # Display information about the dataset
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(data.describe())
```

```python
# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
# Countplot of the 'department' feature
plt.figure(figsize=(10, 6))
sns.countplot(x='department', data=data, hue='is_promoted')
plt.title('Employee Distribution by Department')
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribution of age
plt.figure(figsize=(8, 5))
sns.histplot(data=data, x='age', hue='is_promoted', bins=20)
plt.title('Age Distribution by Promotion')
plt.show()
```

```python
# Pairplot for numerical features
numerical_features = ['no_of_trainings', 'age', 'previous_year_rating', 'length_of_service', 'avg_training_score']
sns.pairplot(data=data, hue='is_promoted', vars=numerical_features)
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```

```python
# Correlation heatmap
correlation_matrix = data[numerical_features].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Countplot of 'education' feature
plt.figure(figsize=(8, 5))
sns.countplot(x='education', data=data, hue='is_promoted')
plt.title('Education Level vs. Promotion')
plt.show()
```

```python
# Countplot of 'gender' feature
plt.figure(figsize=(6, 4))
sns.countplot(x='gender', data=data, hue='is_promoted')
plt.title('Gender vs. Promotion')
plt.show()
```

```python
# Countplot of 'recruitment_channel' feature
plt.figure(figsize=(8, 5))
sns.countplot(x='recruitment_channel', data=data, hue='is_promoted')
plt.title('Recruitment Channel vs. Promotion')
plt.show()
```
